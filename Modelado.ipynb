### Serie de tiempo (modelo ARIMA)

```python
import pandas as pd  
import numpy as np  
from statsmodels.tsa.arima.model import ARIMA  
import matplotlib.pyplot as plt  
  
data = pd.read_csv('localData.csv')
df = data[['YEAR', 'MEDIA', 'ONDA_IRRADIANTE']] 
  
# Ajustar el modelo ARIMA  
model = ARIMA(df['MEDIA'], order=(5,1,0))  
model_fit = model.fit()  

# Predicción para los próximos 50 años  
forecast = model_fit.forecast(steps=50)  

  
# Resumen de los resultados del ajuste  
print(model_fit.summary())  
  
forecast_index = pd.Index(range(df.index[-1]+1, df.index[-1]+51))  # Crear índice para la predicción  

# Gráfica de la serie original y las predicciones  
df['MEDIA'].plot(label='Original')  
model_fit.fittedvalues.plot(color='red', label='Fitted')  
plt.plot(forecast_index, forecast, color='green', label='Forecast') 
plt.legend()  
plt.show()  
```
